### Описание данных

<b>Данные о гипотезах</b>

Файл /datasets/hypothesis.csv. 

* `Hypothesis` — краткое описание гипотезы;
* `Reach` — охват пользователей по 10-балльной шкале;
* `Impact` — влияние на пользователей по 10-балльной шкале;
* `Confidence` — уверенность в гипотезе по 10-балльной шкале;
* `Efforts` — затраты ресурсов на проверку гипотезы по 10-балльной шкале. Чем больше значение Efforts, тем дороже проверка гипотезы.

* `date` — дата, когда был совершён заказ;
* `revenue` — выручка заказа;
* `group` — группа A/B-теста, в которую попал заказ.


Файл /datasets/visitors.csv

* `date` — дата;
* `group` — группа A/B-теста;
* `visitors` — количество пользователей в указанную дату в указанной группе A/B-теста

### Фреймворк ICE

Первым делом определим, как подсчитать ICE score:

$$\text{ICE Score} = \dfrac{\text{Impact} \times \text{Confidence}}{ \text{Efforts}}$$

где: 
* Impact — насколько сильно изменение повлияет на пользователей, их опыт и удовлетворение от продукта;
* Confidence — насколько вы уверены, что это изменение повлияет на пользователей именно так;
* Efforts — сколько стоит протестировать эту гипотезу.


```python
cm = sns.light_palette("green", as_cmap=True)

```

```python
hypothesis.style.background_gradient(cmap=cm)
```

В предоставленных данных все гипотезы уже получили оценки по требуемым параметрам. Поэтому все готово для того, чтобы подсчитать нужное значение 

```python
hypothesis['ICE'] = (hypothesis['Impact'] * hypothesis['Confidence'] / hypothesis['Efforts']).round().astype('int')
```

```python
hypothesis.sort_values(by='ICE', ascending=False)
```

Таким образом, согласно фреймворку ICE, самой приоритетной гипотезой является гипотеза №8: "Запустить акцию, дающую скидку на товар в день рождения". Показатели Impact и Confidence у нее достаточно высоки, а Efforts - небольшой. 

Следущей по приоритетности идет гипотеза под номером 0: "Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей". У нее также очень высоки показатели Confidence и Impact, но Efforts чуть больше, чем в восьмой гипотезе. 

Проверим, останется ли порядок приоритета таким же при работе с фреймворком RICE. 

### Фреймворк RICE

$$\text{ICE Score} = \dfrac{\text{Reach} \times \text{Impact} \times \text{Confidence}}{ \text{Efforts}}$$

где: 
* Reach — скольких пользователей затронет изменение, которое вы хотите внести;
* Impact — насколько сильно изменение повлияет на пользователей, их опыт и удовлетворение от продукта;
* Confidence — насколько вы уверены, что это изменение повлияет на пользователей именно так;
* Efforts — сколько стоит протестировать эту гипотезу.

Главное отличие этого фреймворка от предыдущего - учет количества пользователей, которых затронет изменение. Проверим, как изменился порядок приоритета: 

```python
hypothesis['RICE'] = ((hypothesis['Reach'] * hypothesis['Impact'] * hypothesis['Confidence'] / hypothesis['Efforts'])
                      .round()
                      .astype('int')
                     ) 
```

```python
hypothesis.sort_values(by='RICE', ascending=False).style.background_gradient(subset=['RICE'], cmap=cm)
```

При применении этого фреймворка порядок гипотез изменился достаточно сильно: первая гипотеза фреймворка ICE оказалась на пятом месте, а вот вторая (из ICE) - на третьем. 

На первое место по фреймворку RICE вышла гипотеза "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок". Это звучит разумно, так как показатели Impact и Confidence все еще достаточно высоки, а повлияет это на всех пользователей сайта. 

На втором же месте гипотеза "Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа". У нее низкий Impact, но при этом низкий и Efforts, а вот Reach и Confidence - достаточно велики. 

### Сравнение результатов фреймворков

Как было видно в двух предыдущих шагах, при применении разных фреймворков (хотя они и очень похожи) гипотезы на наиболее приоритетных позициях сильно разнятся. 

Главное отличие фреймворков - в учете параметра Reach - количестве пользователей, на котоых повлияет изменения. Таким образом, анпример, гипотеза об акциях в день рождения, хоть и обладает хорошими показателями, будет в краткосрочном периоде относиться к очень небольшой доле посетителей. При этом при учете параметра Reach на первый план выходит гипотеза о добавлении формы подписки. Она, в свою очередь, затронет всех пользователей интернет-магазина, а значит - с большой вероятностью будет в приоритете.

Посчитайте статистическую значимость различий в конверсии между группами по «сырым» данным. Сделайте выводы и предположения.

Проверять статистическую значимость будем непараметрическим тестом Манна-Уитни, чтобы снизить влияние нетипичных значений. Для этого создадим объекты типа Series с числом заказов каждого из пользователей. 

Создадим Sries следующим образом: 
1. Найдем количество заказов на каждого пользователя 
2. Найдем количество человек, не совершивших ни одного заказа 
3. Заполним Series длинной количества ничего не заказавших людей нолями 
4. Объединим два Series методом pd.concat 

1. Заказы на пользователя 

Данные получим из уже существующей таблицы с количеством заказов на человека 

```python
orders_by_users.head()
```

```python
orders_by_users.query('group == "A"')['orders_by_user']
```

```python
data.head()
```

Таким образом, для пользователей группы А это количество будет следующим:

```python
data['visitors_A'].sum() - orders_by_users.query('group == "A"').shape[0]
```

Для группы В: 

```python
data['visitors_B'].sum() - orders_by_users.query('group == "B"').shape[0]
```

3. Конкатинация Series

```python
sample_A = pd.concat(
                     [
                         orders_by_users.query('group == "A"')['orders_by_user'], 
                         pd.Series(0, 
                                   index=np.arange(
                                       data['visitors_A'].sum() - orders_by_users.query('group == "A"').shape[0]
                                   )
                                  )
                     ]
                    )
```

```python
sample_A
```

```python
sample_B = pd.concat(
                     [
                         orders_by_users.query('group == "B"')['orders_by_user'], 
                         pd.Series(0, 
                                   index=np.arange(
                                       data['visitors_B'].sum() - orders_by_users.query('group == "B"').shape[0]
                                   )
                                  )
                     ]
                    )
```

```python
sample_B
```

<b>Далее найдем статистическую значимость различия</b>

*Сформулируем гипотезы:*
1. Нулевая гипотеза: конверсии групп равны 
2. Альтернативная гипотеза: конверсии групп различны 

Таким образом, любое p-значение больше выбранного значения $\alpha$ будут давать основание отвергнуть нулевую гипотезу и свидетельствовать об отсутствии статистически значимого различия. 

Установим $\alpha = 0,05$ 


Общее количество заказов в сформированной переменной: 

```python
sample_A.sum()
```

То же самое для группы В: 

```python
sample_B.sum()
```

```python
orders.query('group == "B"').shape[0]
```

```python
stats.mannwhitneyu(sample_A, sample_B, alternative = 'two-sided').pvalue
```

P-значение по результатам теста оказалось сильно меньше, чем заявленное alpha. Таким образом, у нас появляются основания отвергнуть нулевую гипотезу о равенстве конверсий в группах. Значит, можно сделать вывод о статистической значимости различий в конверсии
